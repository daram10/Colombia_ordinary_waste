## Data cleaning

### df1 (Recovered)

```python
# IRRELEVANT DATA
# Publicación Aplazada and Sujeto a Verificación* contains data that is not approved by the entity yet, we removed it
df1.drop(df1[df1["TONELADAS APROVECHADAS"] == "Publicación Aplazada"].index, inplace=True)
df1.drop(df1[df1["TONELADAS APROVECHADAS"] == "Sujeto a Verificación*"].index, inplace=True)

# DATES
# Converts date into datetime and months to English. It also creates a new column just with YEAR-MONTH
df1["FECHA"] = pd.to_datetime(df1["FECHA"])

month_dic = {
            "ENERO":"JANUARY",
            "FEBRERO":"FEBRUARY",
            "MARZO":"MARCH",
            "ABRIL":"APRIL",
            "MAYO":"MAY",
            "JUNIO":"JUNE",
            "JULIO":"JULY",
            "AGOSTO":"AUGUST",
            "SEPTIEMBRE":"SEPTEMBER",
            "OCTUBRE":"OCTOBER",
            "NOVIEMBRE":"NOVEMBER",
            "DICIEMBRE":"DECEMBER"
            }

df1["PERIODO"] = df1["PERIODO"].replace(month_dic)
# Adds the column AUX_PER. It is the 'auxiliary period' and contains YEAR-MONTH
df1["AUX_PER"] = pd.to_datetime(df1["ANIO"].astype(str) + df1["PERIODO"], format='%Y%B').dt.to_period("M")

# LOCATION
# Adds the column AUX_LOC. It is the 'auxiliary location' and contains DEPARTMENT-TOWN
# It was made to avoid confusing towns with the same name in different Departments
df1['AUX_LOC'] = df1.apply(lambda x:'%s-%s' % (x['DEPARTAMENTO ORIGEN'],x['MUNICIPIO ORIGEN']),axis=1)

# TYPE
df1["TONELADAS APROVECHADAS"] = df1["TONELADAS APROVECHADAS"].str.replace('[A-Za-z]', '').str.replace(',', '.').astype(float)
```

### df2 (Disposed)

```python
# DATES
# Converts date into datetime and months to English. It also creates a new column just with YEAR-MONTH
df2["PERIODO CARGUE"] = df2["PERIODO CARGUE"].replace(month_dic)
df2["AUX_PER"] = pd.to_datetime(df2["AÑO CARGUE"].astype(str) + df2["PERIODO CARGUE"], format='%Y%B').dt.to_period("M")

#LOCATION
df2['AUX_LOC'] = df2.apply(lambda x:'%s-%s' % (x['DEPARTAMENTO'],x['MUNICIPIO']),axis=1)

# TYPE
df2['TONELADAS DE RESIDUOS'] = df2['TONELADAS DE RESIDUOS'].apply(lambda x: x.replace(",","."))
df2['TONELADAS DE RESIDUOS'] = df2['TONELADAS DE RESIDUOS'].astype(float)
```

```python
# IRRELEVANT DATA
# df3 = df3.dropna()
df3.drop(["Unnamed: 0"],axis=1,inplace=True)

# LOCATION
df3['AUX_LOC'] = df3.apply(lambda x:'%s-%s' % (x['Nombre del departamento'],x['Nombre del municipio']),axis=1)
```

### df4 (Population according to DANE from 1985 to 2017)

```python
# Renames columns to match df5 columns names
df4.rename(columns = {'COD_MUNICIPIO':'DPMP', 'MPNOM':'MPIO'}, inplace = True)

# Removes *total nacional* data from DP column
df4.drop(df4[df4["DP"] == "Total general"].index, inplace=True)

# TYPE
# Converting object to Int pandas
df4['DPMP'] = df4['DPMP'].apply(np.int64)
df4['AÑO'] = df4['AÑO'].apply(np.int64)
df4['DP'] = df4['DP'].astype(str).astype(int)
```

### DANE(Union0): df4 + df5

```python
# Joins df4 and df5
df6 = pd.concat([df4, df5]) #Total population dataset from 1985 to present

# NORMALIZING STRINGS
# Turns lowercase to uppercase
def upper_consistent(df6):
    df6 = df6.apply(lambda x: x.str.upper() if x.dtype == "object" else x) 
    return df6

df6[['DPNOM', 'MPIO']] = upper_consistent(df6[['DPNOM', 'MPIO']])

# Removes accent marks on departamentos and municipios
df6['DPNOM'] = df6['DPNOM'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df6['MPIO'] = df6['MPIO'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

# UNRELEVANT DATA
# Drop NAN values
df6.drop(df6[df6["DPNOM"] == "NaN"].index, inplace=True)
df6.drop(df6[df6["MPIO"] == "NaN"].index, inplace=True)
```

```python
df6["DPNOM"].unique()
```

```python
df6["DPNOM"].replace({'ARCHIPIELAGO DE SAN ANDRES':'ARCHIPIELAGO DE SAN ANDRES, PROVIDENCIA Y SANTA CATALINA'},inplace=True)
df6["MPIO"].replace({'SAN JOSE DE CUCUTA':'CUCUTA'},inplace=True)
df6["MPIO"].replace({'MARIQUITA':'SAN SEBASTIAN DE MARIQUITA'},inplace=True)
df6["MPIO"].replace({'SANTAFE DE ANTIOQUIA':'SANTA FE DE ANTIOQUIA'},inplace=True)

df6['AUX_LOC'] = df6.apply(lambda x:'%s-%s' % (x['DPNOM'],x['MPIO']),axis=1)
df6.rename(columns = {'Total':'Total population'}, inplace = True)
df6.to_csv("data2/DANEnew.csv",sep="|")
df6
```

### TOTwaste (Union1): Recovered + Disposed (TOTAL TONS)

```python
# RECOVERED + DISPOSED TOTAL TONS dataset
# Smaller df1 (location, period and recovered tons sum)
temp1 = df1.groupby(["AUX_LOC","AUX_PER"]).sum()["TONELADAS APROVECHADAS"]
temp1 = temp1.reset_index()

# Smaller df2 (location, period and disposed tons sum)
temp2 = df2.groupby(["AUX_PER","AUX_LOC"]).sum()["TONELADAS DE RESIDUOS"]
# temp2 = df2.groupby(["AUX_LOC","AUX_PER"]).sum()["TONELADAS DE RESIDUOS"]
temp2 = temp2.reset_index()

# New dataset
Union1 = temp2.merge(temp1,on=["AUX_LOC","AUX_PER"],how="left").sort_values(by="TONELADAS APROVECHADAS",ascending=False)
Union1.describe()

# Recovered rate calculation
Union1["Tasa_de_aprovechamiento"] = (Union1["TONELADAS APROVECHADAS"]/(Union1['TONELADAS DE RESIDUOS']+Union1["TONELADAS APROVECHADAS"]))
# Union1.head(2)

# Formula check
Union1[Union1["Tasa_de_aprovechamiento"]==max(Union1["Tasa_de_aprovechamiento"])]
```

```python
# Envigado values were wrong, it was necessary to recalculate and replace them
promedio_envigado =  Union1[(Union1["AUX_LOC"]=="ANTIOQUIA-ENVIGADO") & (Union1["TONELADAS DE RESIDUOS"]!=2.44)].mean()["TONELADAS DE RESIDUOS"]
Union1['TONELADAS DE RESIDUOS'].mask(Union1["TONELADAS DE RESIDUOS"]==2.44, promedio_envigado, inplace=True)
```

```python
# Sets the data range (until 2021-12) and adds the column AÑO
Union1 = Union1[Union1["AUX_PER"]<= "2021-12"]
Union1["AÑO"]= Union1["AUX_PER"].dt.year
Union1.sort_values(by="Tasa_de_aprovechamiento",ascending=False)
Union1.to_csv("data2/TOTwaste.csv",sep="|")
Union1
```

### FULLdf (Union2): Population + Union1

```python
# TOTAL POPULATION
# Smaller df6 (year, total and location) and save .csv
temp6 = df6[df6["ÁREA GEOGRÁFICA"]=="Total"]
temp6 = temp6[["AÑO","Total population","AUX_LOC"]]
# temp6.to_csv("data2/TOTALPopulation.csv",sep="|")
temp6.head(2)
```

```python
# RECOVERED + DISPOSED + POPULATION dataset
Union2 = Union1.merge(temp6,how="left",on=["AUX_LOC","AÑO"])
Union2["Total population"] = Union2["Total population"].apply(lambda x: str(x).replace(".","")).astype(int)
Union2.to_csv("data2/FULLdataset.csv",sep="|")
Union2
```

### data_model: Lifespan + Union2

```python
# RECOVERED + DISPOSED + POPULATION + LIFESPAN dataset
# Smaller Union2
temp7 = Union2.copy()
# # temp7 = temp7.drop(["Tasa_de_aprovechamiento","Total population"],axis=1,inplace=True)
temp7 = temp7.groupby(["AUX_LOC","AÑO"]).sum().reset_index()
temp7

# # New union
Union3 = df6.merge(temp7,on=["AUX_LOC","AÑO"],how="right")
# Union3["Total population_x"] = Union3["Total population"].apply(lambda x: str(x).replace(".",""))
# Union3["Total population"] = Union3["Total population"].astype(int)
Union3.head()

# Smaller Lifespan
temp8 = df3.copy() 
temp8["AÑO"] = 2021
temp8 = temp8[["AUX_LOC","LIFESPAN","Nombre sitio disposición final","AÑO", "LIFESPAN_PERCENTAGE"]]
# temp8

# New dataset
data_model = Union3.merge(temp8,on=["AUX_LOC","AÑO"],how="right")

# # #Recovered rate
data_model["Tasa_de_aprovechamiento"] = (data_model["TONELADAS APROVECHADAS"]/(data_model['TONELADAS DE RESIDUOS']+data_model["TONELADAS APROVECHADAS"]))
data_model["Tasa_de_aprovechamiento"] = data_model["Tasa_de_aprovechamiento"].fillna(0)
data_model
data_model.to_csv("data2/data_model.csv",sep="|")
```
