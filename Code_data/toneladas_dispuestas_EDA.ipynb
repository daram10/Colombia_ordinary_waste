# Data Cleaning

## Data type consistency

NOTAS:

- AÑO CARGUE - INT
- PERIODO CARGUE - DATE

```python
df["AÑO CARGUE"] ## Se puede usar tal cual por ser de tipo int
```

```python
month_dic = {
            "ENERO":"JANUARY",
            "FEBRERO":"FEBRUARY",
            "MARZO":"MARCH",
            "ABRIL":"APRIL",
            "MAYO":"MAY",
            "JUNIO":"JUNE",
            "JULIO":"JULY",
            "AGOSTO":"AUGUST",
            "SEPTIEMBRE":"SEPTEMBER",
            "OCTUBRE":"OCTOBER",
            "NOVIEMBRE":"NOVEMBER",
            "DICIEMBRE":"DECEMBER"
            }

df["PERIODO CARGUE"] = df["PERIODO CARGUE"].replace(month_dic)
df["PERIODO CARGUE"]
```

```python
df['PERIODO_NUEVO'] = pd.to_datetime(df['AÑO CARGUE'].astype(str)  + df['PERIODO CARGUE'], format='%Y%B').dt.to_period("M")
df[["AÑO CARGUE","PERIODO CARGUE","PERIODO_NUEVO"]].head()
```

# Data Augmentation

```python
def label_nacional(x):
    if x == "BOGOTA, D.C.":
        label = "BOGOTA, D.C."
    else:
        label = "Nacional"
    return label
```

```python
df["NACIONAL"]=df["DEPARTAMENTO"].apply(label_nacional)
df
```
