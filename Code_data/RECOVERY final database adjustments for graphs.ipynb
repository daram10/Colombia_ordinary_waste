## Packages

```python
# Imports the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

## Databases

```python
# Call merge databases and others
# FULLdataset (Recovered+Disposed+Population) (it was build in EDA and new dfs)
df1 = pd.read_csv("data2/FULLdataset.csv",sep='|',index_col=0)
# It divides the municipalities in regions
df2 = pd.read_csv("data/Mapas_Traslado_2020.csv",sep=';')
# Full population (it was build in EDA and new dfs)
df3 = pd.read_csv("data2/DANEnew.csv",sep='|',index_col=0)
```

## Cleaning

```python
# DF1
# Drop NAN values from database 
df1.dropna(inplace=True)

#Separate information by "-" from column AUX_LOC
df1[['Departamento','Municipio']]=df1.AUX_LOC.str.split('-',expand=True)
```

```python
# DF2
# Set all values in uppercase
df2['NOM_MUN'] = df2['NOM_MUN'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df2['REGION'] = df2['REGION'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')

#Rename columns name to merge with keyvalue
df2 = df2.rename(columns = {'NOM_MUN':'Municipio'})
```

```python
# DF3
# Adjust values of total population replace by "." and set as an integer
df3["Total population"] = df3["Total population"].apply(lambda x: str(x).replace(".",""))
df3["Total population"] = df3["Total population"].astype(int)
```

### final_data_month

```python
# Merge databases to add variable of interest REGION
Union = df1.merge(df2,on=["Municipio"],how="left")
Union['Tasa_de_aprovechamiento']=Union['Tasa_de_aprovechamiento']*100
Union.shape

# Merge database of population (DANE) with FULL
Union4 = Union.merge(df3,on=["AUX_LOC","AÑO"],how="left")
# Drop unnecessary variables 
Union4.drop(["COD_MUN","TRASLADO DE RECURSOS DE APROVECHAMIENTO","Total population_y"],axis=1,inplace=True)

# Creates some other columns for recovered and disposed waste
Union4['Ton_percapita_apro'] = (Union4['TONELADAS APROVECHADAS']) / (Union4['Total population_x'])
Union4['kg_percapita_apro'] = Union4['Ton_percapita_apro']*1000
Union4['Ton_percapita_noapro'] = (Union4['TONELADAS DE RESIDUOS']) / (Union4['Total population_x'])
Union4['kg_percapita_noapro'] = Union4['Ton_percapita_apro']*1000

#Reorganize order of the variables 
cols = ['AUX_PER', 'AUX_LOC', 'TONELADAS DE RESIDUOS', 'TONELADAS APROVECHADAS',
       'AÑO', 'Total population_x', 'Departamento',
       'Municipio', 'REGION', 'DP', 'DPNOM', 'DPMP', 'MPIO', 'ÁREA GEOGRÁFICA',
       'Tasa_de_aprovechamiento','Ton_percapita_apro', 'kg_percapita_apro', 'Ton_percapita_noapro',
       'kg_percapita_noapro']
Union4 = Union4[cols]

#Exports information 
Union4.to_csv("graphs/RECOVERED_MAPS/data/final_data_month.csv",sep="|")
```

### final_data_year

```python
# Creates a copy of the "Union4" database
temp0 = Union4
temp0

# Group information by region, department, municipality and year
temp1 = temp0.groupby(["REGION", "AUX_LOC","AÑO"]).sum().reset_index()
temp1

#Merge database of population (DANE) with waste information base (SSPD)
Union3 = df3.merge(temp1,on=["AUX_LOC","AÑO"],how="right")

#Separate information by "-" from column AUX_LOC
Union3[['Departamento','Municipio']]=Union3.AUX_LOC.str.split('-',expand=True)

#Reorganize order of the variables
#cols = Union3.columns.tolist()
cols = ['AÑO', 'REGION', 'AUX_LOC', 'Departamento', 'Municipio', 'Total population',  'TONELADAS DE RESIDUOS', 'TONELADAS APROVECHADAS']
Union3 = Union3[cols]

#Create "rate for the used of ordinary solid waste" variable
Union3["Tasa de aprovechamiento"] = (Union3["TONELADAS APROVECHADAS"]/(Union3["TONELADAS DE RESIDUOS"]+Union3["TONELADAS APROVECHADAS"])) * 100

#Create "Tons effectively used per capita" variable
Union3['Ton_percapita_apro'] = (Union3['TONELADAS APROVECHADAS']) / (Union3['Total population'])
Union3['kg_percapita_apro'] = Union3['Ton_percapita_apro']*1000
Union3['Ton_percapita_noapro'] = (Union3['TONELADAS DE RESIDUOS']) / (Union3['Total population'])
Union3['kg_percapita_noapro'] = Union3['Ton_percapita_noapro']*1000

#Export information 
Union3.to_csv("graphs/RECOVERED_MAPS/data/final_data_year.csv",sep="|")
```
