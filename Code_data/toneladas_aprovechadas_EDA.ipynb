```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Imports the dataset, this is the same 'toneladas_aprovechadas_mayo22.csv' but with less rows because it was an earlier month
df = pd.read_csv("data/toneladas_aprovechadas.csv")
```

```python
df.head()
```

# Data Cleaning

## Relevance

```python
df.drop(df[df["TONELADAS APROVECHADAS"] == "Publicación Aplazada"].index, inplace=True)
df.drop(df[df["TONELADAS APROVECHADAS"] == "Sujeto a Verificación*"].index, inplace=True)
```

## Data type consistency

| Column | Current data type | Convert to |
| --- | --- | --- |
| ID EMPRESA (Optional)| `int64` | `category` |
| ANIO | `int64` | `datetime` |
| PERIODO | `object` | `datetime` |
| FECHA | `object` | `datetime` |
| TONELADAS APROVECHADAS | `object` | `float64` |


```python
df["FECHA"] = pd.to_datetime(df["FECHA"])
```

```python
df["TONELADAS APROVECHADAS"] = df["TONELADAS APROVECHADAS"].astype(float)
```

# Data Augmentation

```python
df.info()
```

```python
month_dic = {
            "ENERO":"JANUARY",
            "FEBRERO":"FEBRUARY",
            "MARZO":"MARCH",
            "ABRIL":"APRIL",
            "MAYO":"MAY",
            "JUNIO":"JUNE",
            "JULIO":"JULY",
            "AGOSTO":"AUGUST",
            "SEPTIEMBRE":"SEPTEMBER",
            "OCTUBRE":"OCTOBER",
            "NOVIEMBRE":"NOVEMBER",
            "DICIEMBRE":"DECEMBER"
            }

df["PERIODO"] = df["PERIODO"].replace(month_dic)
df["PERIODO_NUEVO"] = pd.to_datetime(df["ANIO"].astype(str) + df["PERIODO"], format='%Y%B').dt.to_period("M") ## CAMBIAR NOMBRE VARIALBLE
```

```python
def label_nacional(x):
    if x == "BOGOTA, D.C.":
        label = "BOGOTA, D.C."
    else:
        label = "NACIONAL"
    return label
df["NACIONAL"]=df["DEPARTAMENTO ORIGEN"].apply(label_nacional)
```

```python
df.to_csv("data/toneladas_aprovechadas_clean.csv")
```

# Basic EDA

```python
numero_reportes = df.groupby(df["PERIODO_NUEVO"]).size()

numero_toneladas_2020 = df[(df["PERIODO_NUEVO"] >= "2020-01") & (df["PERIODO_NUEVO"] <= "2020-12")].groupby(["PERIODO_NUEVO"]).sum()["TONELADAS APROVECHADAS"
                                                                                                                                    ]
toneladas_departamentos = df.groupby(df["DEPARTAMENTO ORIGEN"]).sum()["TONELADAS APROVECHADAS"].sort_values(ascending=False).head(28) # OMITI EL ULTIMO VALOR

toneladas_anuales_nacional = df.groupby(["ANIO","NACIONAL"]).sum()["TONELADAS APROVECHADAS"].unstack(level = 1)

toneladas_mensuales = df[df["ANIO"] <= 2020].groupby(["PERIODO","ANIO"]).sum()["TONELADAS APROVECHADAS"].unstack(level = 1)
month_order = ["JANUARY", "FEBRUARY", "MARCH", "APRIL", "MAY", "JUNE", "JULY", "AUGUST", "SEPTEMBER", "OCTOBER", "NOVEMBER", "DECEMBER"]
toneladas_mensuales = toneladas_mensuales.reindex(month_order, axis=0)

toneladas_empresas = df.groupby(df["EMPRESA"]).sum()["TONELADAS APROVECHADAS"].sort_values(ascending=False).head(10)
```

## Número de reportes

```python
numero_reportes.plot(figsize=(16,9), title="Number of reports per period")
plt.xlabel('Period')
plt.ylabel('Total of reports')
```

## Toneladas mensuales

```python
toneladas_mensuales.plot.bar(subplots = True, figsize=(18,30), title="Tons disaggregated by month and year")
plt.xlabel('Month')
plt.ylabel('Total of tons')
plt.xticks(rotation='horizontal')
```

```python
toneladas_mensuales.plot.bar(subplots = False, figsize=(16,9), title="Tons disaggregated by month and year")
plt.xlabel('Month')
plt.ylabel('Total of tons')
plt.xticks(rotation='horizontal')
plt.legend(title='Year')
```

## Toneladas 2020 (ultimo año con reporte completo)

```python
numero_toneladas_2020.plot.bar(figsize=(16,9), title="Number of tons for 2020")
plt.xlabel('Month')
plt.ylabel('Total of tons')
plt.xticks(rotation='horizontal')
```

## Toneladas por departamento

```python
toneladas_departamentos.plot.bar(figsize=(16,9), title="Total tons reported by department")
plt.xlabel('Department of origin')
plt.ylabel('Total of tons')
```

## Toneladas anuales  diferenciado entre Nacional y Bogotá

```python
toneladas_anuales_nacional.plot.bar(figsize=(10,7), title="Reused waste  Bogota - other departments", stacked= True)
plt.xlabel('Year')
plt.ylabel('Total of tons')
plt.xticks(rotation='horizontal')
plt.legend(title='')
plt.rcParams.update({'font.size': 15})
```

## Top 10 de las empresas que más toneladas reciclan

```python
toneladas_empresas.plot.bar(figsize=(16,9), title="Top 10 of companies which report more tons")
plt.xlabel('Company')
plt.ylabel('Total of tons')
```
